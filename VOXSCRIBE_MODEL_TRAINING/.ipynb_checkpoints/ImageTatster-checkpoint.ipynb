```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Loading the trained model
model = load_model("best_model.h5")
print("Model is loaded successfully!")

def preprocess_image(image_path):

    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    resized_image = cv2.resize(image, (28, 28))
    inverted_image = 255 - resized_image
    normalized_image = inverted_image / 255.0
    reshaped_image = normalized_image.reshape(1, 28, 28, 1)
    return reshaped_image


image_path = "testimg5.png"
# preparing the image
processed_image = preprocess_image(image_path)

# Predicting the character
prediction = model.predict(processed_image)
predicted_class = np.argmax(prediction)

# Decoding the predicted class
def decode_prediction(index):
    if 0 <= index <= 9:
        return str(index)  
    elif 10 <= index <= 35:
        return chr(index - 10 + ord('A'))  
    elif 36 <= index <= 61:
        return chr(index - 36 + ord('a')) 
    else:
        return "Unknown"

decoded_character = decode_prediction(predicted_class)
print(f"Predicted Character: {decoded_character}")

# Display the input image
plt.imshow(cv2.imread(image_path, cv2.IMREAD_GRAYSCALE), cmap="gray")
plt.title(f"Predicted Character: {decoded_character}")
plt.axis("off")
plt.show()
```
